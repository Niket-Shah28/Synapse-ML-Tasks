```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("/content/gdrive/MyDrive/Task2hp/HPSpells.csv")
```

```python
data.head()
```

```python
data.drop(data[data.Type == 'Hex'].index,inplace=True)
```

```python
data
```

```python
data.dropna(subset = ['Light'],inplace = True)
```

```python
data
```

```python
Types = data.Type.unique()
count={}
for Type in Types:
  i=0
  for item in data.Type:
    if item == Type:
      i=i+1
  count[Type]=i

plt.figure(figsize=(10,7))
plt.bar(range(len(count)),list(count.values()))
plt.xticks(range(len(count)),Types)
plt.xlabel('Spells')
plt.ylabel('Frequency of Spells')
plt.title('Spell Distribution')
plt.show()
```
